```python
# Simple Markov Chain

import matplotlib.pyplot as plt
import numpy as np

S = [1, 2, 3, 4] # state space
P = [[0, 0.2, 0.4, 0.4],  # transition matrix
     [0.9, 0.1, 0, 0], 
     [0.5, 0.2, 0.2, 0.1],
     [0.1, 0.2, 0.3, 0.4]]

init_state = 1 # initial state value
T = 50 # simulation periods

# Arrays
S = np.array(S)
P = np.array(P)
H = np.empty(T+1)
H[0] = init_state

# Simulations
for t in range(T):
    H[t+1] = np.random.choice(S, p = P[np.where(S == H[t])][0])
plt.plot(H, 'go--')
plt.show()

for t in range(T):
    H[t+1] = np.random.choice(S, p = P[np.where(S == H[t])][0])
plt.plot(H, 'go--')
plt.show()
```

```python
# Large State Markov Chain

import matplotlib.pyplot as plt
import numpy as np

S = [1, 2, 3, 4, 5, 6] # state space
P = [[0, 0.2, 0.4, 0.4, 0, 0],  # transition matrix
     [0, 0, 0.9, 0.1, 0, 0], 
     [0, 0.5, 0.2, 0.2, 0.1, 0],
     [0, 0.1, 0.2, 0.3, 0.4, 0], 
     [0.1, 0.1, 0.1, 0.2, 0.2, 0.3], 
     [0.9, 0.1, 0, 0, 0, 0]]

init_state = 1 # initial state value
T = 100 # simulation periods

# Arrays
S = np.array(S)
P = np.array(P)
H = np.empty(T+1)
H[0] = init_state

# Simulations
for t in range(T):
    H[t+1] = np.random.choice(S, p = P[np.where(S == H[t])][0])
plt.plot(H, 'go--')
plt.show()

for t in range(T):
    H[t+1] = np.random.choice(S, p = P[np.where(S == H[t])][0])
plt.plot(H, 'go--')
plt.show()
```
